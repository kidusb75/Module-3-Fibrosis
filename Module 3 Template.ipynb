## Project Goal:
This project seeks to analyze lung tissue images to determine how the extent of fibrosis changes with depth into the lung. By quantifying the percentage of white pixels at different depths, we aim to develop an interpolation-based model that predicts fibrosis at depths where direct measurements are unavailable. This helps simulate how a medical device could estimate fibrosis without requiring biopsies at every depth.

## Disease Background: 

* Prevalence & incidence

    Pulmonary fibrosis is a chronic lung disease that affects millions of people worldwide, with Idiopathic Pulmonary Fibrosis (IPF) being the most common and well-studied form. Globally, pulmonary fibrosis is estimated to affect approximately 3 to 5 million people. The prevalence of IPF specifically is about 13 to 20 cases per 100,000 individuals, with the United States having an estimation between 100,000 and 200,000 people that are currently living with the disease. It's more common in older adults, particularly those over the age of 60, and occurs more frequently in men than in women (Meltzer & Noble, 2008; MedlinePlus Genetics, 2020).

    In terms of incidence, IPF is considered a rare disease but continues to be diagnosed more frequently. Globally, there are about 3 to 9 new cases per 100,000 people each year, and in the United States, this corresponds to approximately 30,000 to 40,000 new cases annually. The rise in new cases is likely due to an aging population, improved ability to diagnose the disease, and increased awareness among healthcare providers (Meltzer & Noble, 2008; MedlinePlus Genetics, 2020).



* Risk factors (genetic, lifestyle)

    The likelihood of developing idiopathic pulmonary fibrosis (IPF) rises with age, particularly in individuals in their 60s and 70s. Men are more commonly affected than women, and lifestyle habits such as smoking can further elevate the risk. A family history of IPF can also increase risk, as having a parent or sibling with the condition makes one more susceptible. Furthermore, specific genetic mutations in over ten different genes have been associated with a greater chance of developing the disease (National Heart, Lung, and Blood Institute, 2022).

* Symptoms

    Idiopathic pulmonary fibrosis (IPF) most often causes shortness of breath that slowly gets worse, beginning with activity and eventually happening even when at rest, and a dry cough that can occur frequently and be difficult to control. Other symptoms may include muscle and joint aches, clubbing of the fingers or toes where the tips become widened and rounded, severe fatigue, unintentional weight loss, a general feeling of illness, and fast, shallow breathing (National Heart, Lung, and Blood Institute, 2022).


* Standard of care treatment(s)

    The standard approach to managing pulmonary fibrosis aims to slow the progression of the disease, ease symptoms, and enhance quality of life. Core treatment usually includes antifibrotic drugs like nintedanib or pirfenidone, which help limit further lung scarring. Patients who experience low oxygen levels may require supplemental oxygen to support breathing and daily activities. Pulmonary rehabilitation, which incorporates exercise, breathing exercises, and patient education, is recommended to preserve lung function and overall well-being. Additional care includes lifestyle modifications such as smoking cessation, avoiding lung irritants, and maintaining good nutrition, along with symptom management for cough or acid reflux. For patients with advanced or progressive disease, lung transplantation may be considered when standard therapies are no longer effective (Temple Health, 2022)


* Biological mechanisms (anatomy, organ physiology, cell & molecular physiology)

    Lung fibrosis, including idiopathic pulmonary fibrosis (IPF), happens when the lungs gradually develop excess scar tissue, which makes it harder for them to work properly. Normally, the lungs are made up of branching airways and tiny air sacs called alveoli, where oxygen passes into the blood and carbon dioxide leaves. The walls of the alveoli are very thin and made of alveolar epithelial cells (type I and II) and small blood vessels. Type I cells cover most of the alveoli and help gases move efficiently, while type II cells produce surfactant to keep the lungs open and can repair the alveolar lining after injury. When these cells are repeatedly damaged by infections, toxins, or unknown causes, the structure of the lungs becomes abnormal and scar tissue starts to build up (Zhang, Y., & Wang, J., 2023).

    Normally, after a lung injury, the body triggers inflammation to heal the tissue. In lung fibrosis, however, this process becomes uncontrolled. Immune cells and injured lung cells release signals, especially transforming growth factor‑β (TGF‑β), which activate fibroblasts. These fibroblasts turn into myofibroblasts, cells that produce large amounts of extracellular matrix (ECM), including collagen. Too much ECM makes the lungs stiff and thickens the walls of the alveoli, which reduces lung compliance and makes it harder for oxygen to enter the blood (Agarwal, A. K., & Huda, N., 2023).

    At the cellular level, the problem is that lung cells cannot repair themselves normally. Some alveolar cells die, while others release chemicals that encourage fibroblasts to grow and produce more ECM. Processes like epithelial‑to‑mesenchymal transition (EMT) may also convert some epithelial cells into fibroblast‑like cells. Key signaling pathways, such as TGF‑β/Smad and Wnt/β‑catenin, drive this abnormal cell activity and further promote scarring (Grafanaki et al., 2025).

    Finally, a major issue in lung fibrosis is the failure to resolve the scar tissue. In healthy repair, fibroblasts eventually stop producing ECM and die off, and enzymes break down excess matrix. In fibrotic lungs, this does not happen. Myofibroblasts remain active, ECM continues to accumulate, and the lung tissue becomes permanently stiff and thickened. This structural change causes the main symptoms of fibrosis: shortness of breath, reduced exercise tolerance, and low blood oxygen (Zhang, Y., & Wang, J., 2023).

Unpublished data was collected by the Peirce-Cottler Lab (Dept. of Biomedical Engineering) and Kim Lab (Division of Pulmonary and Critical Care) at the University of Virginia School of Medicine. The data was derived using a Bleomycin-induced Lung Injury Model, a standard experimental approach where the antibiotic Bleomycin is used to simulate the progression of Idiopathic Pulmonary Fibrosis (IPF) in mice. Following a three-week induction period, lung tissue from the upper, middle, and lower lobes was digitally imaged. The resulting dataset includes 78  images captured at specific biopsy depths. To quantify the severity of the disease, we analyzed these images to measure the percentage of white pixels, a computational proxy for fibrotic collagen deposition—relative to the total tissue area. All depth measurements are recorded in microns (µm).

## Limitations and Future Work:

A primary limitation of this project is the relatively small number of data points used for interpolation, which may not capture the non-linear biological complexity of human lung tissue. Furthermore, mouse models are approximations of human disease and may not perfectly replicate the spatial distribution of human IPF.

If given more time, we would implement automated image segmentation using convolutional neural networks (CNNs) to distinguish between collagen and other bright artifacts (e.g., shadows or blood vessels).
